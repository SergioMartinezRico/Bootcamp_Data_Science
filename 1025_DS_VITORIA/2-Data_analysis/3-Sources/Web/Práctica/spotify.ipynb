metodo -> sacar Id de artista -> sacar los albumes y sus id -> iterar por cada id de album y sacar las canciones

```python
#buscar el ID del artista

busqueda = sp.search(q="Sidonie", type="artist", limit=1)
print("la busqueda es de tipo:")
print (type(busqueda))
print (busqueda.keys())
print("-------------------------------")
#print(json.dumps(busqueda, indent=4, ensure_ascii=False))
artistas = busqueda["artists"]
print("artistas es de tipo:")
print (type(artistas))
print (artistas.keys())
print("-------------------------------")
#print(json.dumps(artistas, indent=4, ensure_ascii=False))
artista = artistas['items'][0] #en la busqueda coge el primer valor
print("-------------------------------")
print("artista es de tipo:")
print (type(artista))
print(artista.keys())
identificador_art = artista['id'] #del primer valor coge la clave id
print(identificador_art)
```

```python
#con el ID del artista buscar los albumes
albums = sp.artist_albums(
    artist_id=identificador_art,
    album_type='album',
    limit=50,
    country='ES'
)
print (albums.keys())
lista_albums = albums["items"]

print(len(lista_albums))
print (type(albums["items"]))
#print(json.dumps(lista_albums, indent=4, ensure_ascii=False)) lo uso para ver lo que devuelve
```

```python
canciones_sidonie=[] #lista de canciones vacia

for album in lista_albums: #por cada album en la lista tenemos album[name] y album[id] que sale de la lista_albums anterior
  #print(type(album))
  #print(album.keys()) para ver sus claves
  titulo_album = album["name"]
  identificador_alb = album["id"]
  canciones = sp.album_tracks (album_id=identificador_alb)  #con sp.album_tracks se sacan las canciones por cada album id
  #print(type(canciones)) con este print vemos que es un diccionario. Con la clave items sacamos cada cancion
  #print(canciones["items"]) 
  for cancion in canciones["items"]:  #por cada cancion en la lista creamos un diccionario de artista, album, cancion
    #print (cancion["name"])
    canciones_sidonie.append({
        "Artista": "Sidonie",
        "Álbum": titulo_album,
        "Canción": cancion["name"]
    })
print (canciones_sidonie)
```

```python
import pandas as pd 
df = pd.DataFrame(canciones_sidonie)
df
```
