2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0,5,50)
recta_1 = 3*x
recta_2 = -3*x -5
#ej del 1 al 4
fig = plt.figure()
plt.subplot(1,2,1)
plt.plot(x,recta_1,"-")
plt.xlim(0,5)
plt.ylim(-15,15)
plt.title("Recta sube")
plt.ylabel("eje Y")
plt.xlabel("eje X")
plt.subplot(1,2,2)
plt.plot(x,recta_2,"--")
plt.title("Recta baja")
plt.xlim(0,5)
plt.ylim(-15,15)
plt.xlabel("eje X")
```

10. Prueba a hacer el gráfico con las barras en horizontal.

```python
alturas = [160, 180, 175, 190, 170]
nombres =["Ana", "Luis", "Pedro", "Sofía","Carmen"]
plt.figure(figsize=(3.5,3.5)) #llamada a figure antes del grafico
plt.barh(nombres, alturas, color="green")
plt.title("Alturas (Barras Horizontales)")
plt.xlabel("Altura (cm)")
plt.show()
```

11. Prueba a invertir los ejes con ax.invert_axis()
Necesitarás tener los ejes disponibles fuera de la figura

```python
alturas = [160, 180, 175, 190, 170]
nombres = ["Ana", "Luis", "Pedro", "Sofía", "Carmen"]

# Usar subplots para obtener los ejes (ax) por separado
fig, ax = plt.subplots(figsize=(5, 3.5)) 
ax.barh(nombres, alturas, color="green")
ax.set_title("Alturas (Eje Y Invertido)")
ax.set_xlabel("Altura (cm)")
ax.invert_xaxis() # Invierte el eje Y para que 'Ana' esté arriba

plt.show()
```

13. Vamos a introducir pandas. Crea un DataFrame con las columnas "Year" de valores 2015, 2016, 2017, 2018, 2019 y la columna "Sold_items_A" de valores 1000, 3500, 4000, 5500, 7000

```python

data = {
    "Year": [2015, 2016, 2017, 2018, 2019],
    "Sold_items_A": [1000, 3500, 4000, 5500, 7000]
}
df = pd.DataFrame(data)

```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
plt.style.use('seaborn-v0_8-white') # Cambia el estilo a 'seaborn-white'

plt.figure(figsize=(6, 4))
# Pinta la línea con rayas y color verde ('--g')
plt.plot(df["Year"], df["Sold_items_A"], '--g', label="Departamento A") 
plt.title("Ventas Departamento A por Año")
plt.xlabel("Año")
plt.ylabel("Unidades Vendidas")
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
df["Sold_items_B"] = [2000, 3100, 5000, 4000, 6000] # Añade la columna B

plt.style.use('seaborn-v0_8-white')

plt.figure(figsize=(6, 4))
# Departamento A: línea con rayas y verde ('--g')
plt.plot(df["Year"], df["Sold_items_A"], '--g', label="Departamento A")
# Departamento B: línea punteada y roja (':r')
plt.plot(df["Year"], df["Sold_items_B"], ':r', label="Departamento B")

plt.title("Ventas Departamentos A y B por Año")
plt.xlabel("Año")
plt.ylabel("Unidades Vendidas")
plt.legend(loc="lower right") # Muestra la leyenda abajo a la derecha
plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

```python
np.random.seed(8) 
alturas_aleatorias = np.random.randint(150, 191, 300)
df_alturas = pd.DataFrame({'Altura': alturas_aleatorias})
df_alturas.plot(kind='hist', bins=20, title='Distribución de 300 Alturas Aleatorias', xlabel='Altura (cm)', ylabel='Frecuencia',legend=False)
plt.show()
```
