```python
import psycopg2
import pandas as pd
```

```python
hosting = "localhost"
puerto = "5432"
nombre_db = "Proyecto_DB_Bootcamp"
usuario = "admin"
pswd = "Noa190219"
```

```python
conn = psycopg2.connect(
        host = hosting,
        port = puerto,
        dbname = nombre_db,
        user = usuario,
        password = pswd
)
```

```python
cursor = conn.cursor()
```

```python
def sql_query(query):   # SELECT * FROM TRACKS

    # Ejecuta la query
    cursor.execute(query)

    # Almacena los datos de la query 
    ans = cursor.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cursor.description]

    return pd.DataFrame(ans,columns=names)
```

```python
consulta = "SELECT id_libro, titulo, nombre, nacionalidad FROM libro INNER JOIN autor ON libro.id_autor = autor.id_autor WHERE nombre = 'Gabriel García Márquez'"
```

```python
cursor.execute(consulta)
```

```python
libros = cursor.fetchall()
```

```python
print(libros)
```

```python
df_libros_gabriel = sql_query(consulta)
df_libros_gabriel
```
